### One Hot Encoding with many categories in a feature

```python
df=pd.read_csv('mercedesbenz.csv')
df.head()
```

```python
df=pd.read_csv('mercedesbenz.csv',usecols=["X0","X1","X2","X3","X4","X5","X6"])
```

```python
df.head()
```

```python
## printing all the categories
for features in df.columns:
    print(df[features].value_counts())
```

```python
df['X0'].unique()
```

```python
for features in df.columns:
    print(f'{features} having {len(df[features].unique())} category')
```

```python
df.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
df.X1.value_counts().sort_values(ascending=False).head(10).index
```

```python
list_10=df.X1.value_counts().sort_values(ascending=False).head(10).index
list_10=list(list_10)
list_10
```

```python
for categories in list_10:
    df[categories]=np.where(df['X1']==categories,1,0)
```
